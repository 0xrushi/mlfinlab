Universal Portfolio

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlfinlab.online_portfolio_selection import *

```

```python
stock_prices = pd.read_csv('../../tests/test_data/stock_prices.csv', parse_dates=True, index_col='Date')
stock_prices = stock_prices.dropna(axis=1)
```

```python
up10 = UP(number_of_experts=10)
up10.allocate(stock_prices, resample_by='M')
up10000 = UP(number_of_experts=10000)
up10000.allocate(stock_prices, resample_by='M')
```

```python
up10.all_weights
up10000.all_weights
```

```python
plt.figure(figsize=(20,10))
plt.title('UP with 10 Experts', fontsize=40)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Portfolio Returns', fontsize=20)
plt.plot(up10.portfolio_return)
plt.savefig('up10.png')
```

```python
plt.figure(figsize=(20,10))
plt.title('UP with 10000 Experts', fontsize=40)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Portfolio Returns', fontsize=20)
plt.plot(up10000.portfolio_return)
plt.savefig('up10000.png')

```
